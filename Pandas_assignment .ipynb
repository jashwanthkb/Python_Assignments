```python
 #Calculate Total Sales Amount for each transaction using a new column
df["Total Sales Amount"]=(df["Quantity"]*df["UnitPrice"])*(1-df["Discount"]/100)
df
```

```python
#Find total sales by each Region.
df["Total Sales Amount"].groupby(df["Region"]).sum()
```

```python
#Display all transactions where Discount is greater than 5%.
df[df["Discount"]>5]
```

```python
#Group data by Category and find the average UnitPrice for each.
df["UnitPrice"].groupby(df["Category"]).mean()
```

```python
#Find the top 2 SalesPersons with the highest total sales amount.

df["Total Sales Amount"].groupby(df["SalesPerson"]).sum().nlargest(2)
```

```python
#Filter records for only Electronics category where sales are above ₹50,000.

df[(df["Category"]=="Electronics") & (df["Total Sales Amount"]>50000)]
```

```python
#Sort the dataset by Date and SalesPerson in ascending order.
df.sort_values(["Date","SalesPerson"], ascending=True)
```

```python
#Find the number of transactions for each CustomerType (Regular vs New).
df["CustomerType"].groupby(df["CustomerType"]).count()
```

```python
#Get the most frequently sold product in each region.
df.groupby(["Region"])["Product"].agg(lambda x: x.mode()[0])
```

```python
#Pivot Table: Create a pivot table showing total sales by Region and Category.
pd.pivot_table(df,values=["Total Sales Amount"],index=["Region"],columns=["Category"],aggfunc="sum")
```
