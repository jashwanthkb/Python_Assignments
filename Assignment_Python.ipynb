```python
color=input("Enter the color name:")
color_n=color.lower()
if (color_n=="red"):
    print("STOP")
elif(color_n=="yellow"):
    print("Ready to Go")
elif(color_n=="green"):
    print("GO")
else:
    print("invalid Color")
```

```python
Marks=int(input("Enter the Students Marks:"))
if (Marks>100 or Marks<0):
    print("Invalid marks")
elif(Marks>=90):
    print("Grade A")
elif(Marks>=75 and Marks<90):
    print("Grade B")
elif(Marks>=50):
    print("Grade C")
else:
    print("Fail")
```

```python
num=int(input("Enter the Number:"))
if(num%2==0 and num%5==0):
    print(f"the number {num} is even and divisible by 5")
elif(num%2==0):
    print("The number is even but not divisible by 5")
elif(num%5==0):
    print("The number is odd but divisible by 5")
else:
    print("The number is odd and not divisible by 5")
```
